```python
import pandas as pd
import sys
import os
import csv
import re

base_path_data = os.path.join(os.path.abspath(''), 'data')
base_path_articles = os.path.join(os.path.abspath(''), 'articles')
```

```python
dnn_articles_2000_2020 = pd.read_csv(os.path.join(base_path_data, 'dnn_articles_2000_2020.csv'), index_col=0)
dnn_articles_2022_2024_1 = pd.read_csv(os.path.join(base_path_data, 'dnn_articles_2022_2024_1.csv'),index_col=0)
dnn_articles_2022_2024 = pd.read_csv(os.path.join(base_path_data, 'dnn_articles_2022_2024.csv'),index_col=0)
```

```python
dnn_articles_2022_2024_1
```

```python
dnn_articles = pd.concat([dnn_articles_2000_2020, dnn_articles_2022_2024_1,dnn_articles_2022_2024], ignore_index=True)
```

```python
dnn_articles.drop_duplicates(subset=['title','date'], inplace=True)
```

```python
dnn_articles["source"] = "DNN"
```

```python
dnn_articles
```

```python
dnn_articles.to_csv(os.path.join(base_path_articles, 'dnn_articles.csv'),quoting=csv.QUOTE_NONNUMERIC,index_label='doc_id')
```

# Süddeutsche

```python
sz = []
for file in os.listdir("data/"):
    finding = re.findall(r'sz_articles.*', file)
    if finding:    
        sz.append(finding[0])
sz
```

```python
complete_sz = pd.DataFrame(columns=['title','author','category', 'date', 'text'])
for article in sz:
    iteration = pd.read_csv(os.path.join(base_path_data, article), index_col=0)
    complete_sz = pd.concat([complete_sz, iteration],ignore_index=True)
complete_sz
```

```python
complete_sz.drop_duplicates(subset=['title','date'], inplace=True)
```
